```python
# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Get the list of stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Function to perform lemmatization on a word
def lemmatize_word(word):
    pos = get_wordnet_pos(word)  # Get the POS tag for the word
    if pos is None:
        return lemmatizer.lemmatize(word)
    else:
        return lemmatizer.lemmatize(word, pos=pos)

# Function to map POS tags from the WordNet lemmatizer to the NLTK POS tags
def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {
        "J": wordnet.ADJ,
        "N": wordnet.NOUN,
        "V": wordnet.VERB,
        "R": wordnet.ADV
    }
    return tag_dict.get(tag, wordnet.NOUN)

# Preprocessing function
def preprocess_text(text):
    # Remove words starting with '@' symbol
    text = re.sub(r'@\w+\s?', '', text)
    
    # Remove words starting with 'http://'
    text = re.sub(r'http://\S+', '', text)
    
    # Remove special characters
    text = re.sub(r'[^\w\s]', '', text)
    
    # Lowercase the text
    text = text.lower()
    
    # Tokenize the text into words
    words = nltk.word_tokenize(text)
    
    # Lemmatize the words
    lemmatized_words = [lemmatize_word(word) for word in words]
    
    # Remove stopwords
    filtered_words = [word for word in lemmatized_words if word not in stop_words]
    
    # Join the filtered words back into a single string
    preprocessed_text = ' '.join(filtered_words)
    
    return preprocessed_text

start_time = time.time()
df['text'] = df['text'].apply(preprocess_text)
end_time = time.time()

print("Preprocessing execution Time: {:.2f} seconds".format(end_time - start_time))
```

Naive Bayes (Default alpha value = 1)

```python
dfs = [df_sampled, df_full]
accuracy_scores = []

for df_name, df in zip(['df_sampled', 'df_full'], dfs):
    print(f"Processing {df_name}...")
    
    # Split the data into features (X) and target (y)
    X = df['text']
    y = df['target']

    # Encode the text column using Bag of Words representation
    start_time_vectorize = time.time()
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(X)
    vectorizing_time = time.time() - start_time_vectorize

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.10, random_state=42)

    alpha_values = np.linspace(0, 3, 5)  # Generate 5 values between 0 and 1
    accuracy_scores_df = []

    for alpha in alpha_values:
        # Initialize the Naive Bayes classifier with the current alpha value
        naive_bayes = MultinomialNB(alpha=alpha)

        # Train the classifier
        start_time_training = time.time()
        naive_bayes.fit(X_train, y_train)
        training_time = time.time() - start_time_training

        # Make predictions on the test set
        y_pred = naive_bayes.predict(X_test)

        # Calculate accuracy and append it to the list
        accuracy = accuracy_score(y_test, y_pred)
        accuracy_scores_df.append(accuracy)

        # Print the classification report
        print(f"Classification Report (DataFrame {df_name}, alpha={alpha}):")
        print(classification_report(y_test, y_pred))

        # Print the confusion matrix
        print(f"Confusion Matrix (DataFrame {df_name}, alpha={alpha}):")
        print(confusion_matrix(y_test, y_pred))

        # Print the running time
        print("Vectorizing Time: {:.2f} seconds".format(vectorizing_time))
        print("Training Time: {:.2f} seconds".format(training_time))
        print("-----------------------------------")
    
    accuracy_scores.append(accuracy_scores_df)
    
# Plot accuracy per model
alpha_labels = [f'alpha={alpha}' for alpha in alpha_values]
for i, accuracy_scores_df in enumerate(accuracy_scores):
    df_name = ['df_sampled', 'df_full'][i]
    plt.plot(alpha_labels, accuracy_scores_df, label=df_name)

plt.xlabel('Alpha')
plt.ylabel('Accuracy')
plt.title('Accuracy per Model')
plt.legend()
plt.show()
```

```python
tfhub_handle_encoder = 'https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/3'
tfhub_handle_preprocess = 'https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3'
```

```python
def build_classifier_model():
    text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
    preprocessing_layer = hub.KerasLayer(tfhub_handle_preprocess, name='preprocessing')
    encoder_inputs = preprocessing_layer(text_input)
    encoder = hub.KerasLayer(tfhub_handle_encoder, trainable=True, name='BERT_encoder')
    outputs = encoder(encoder_inputs)
    net = outputs['pooled_output']
    net = tf.keras.layers.Dropout(0.1)(net)
    net = tf.keras.layers.Dense(1, activation=None, name='classifier')(net)
    return tf.keras.Model(text_input, net)
```
